```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 1

```python
#a) 
df = pd.read_csv("/Users/emiliodulay/Documents/Programming/Math42/alcohol_induced_deaths.csv")
df["Ten-Year Age Groups"] = df["Ten-Year Age Groups"].str.replace(" years", "", regex=False)
df.head()
df.tail()
```

```python
#b)


plt.plot(df[df["Ten-Year Age Groups"] == "45-54"]["Year"], df[df["Ten-Year Age Groups"] == "45-54"]["Crude Rate"], label = "45-54", marker = ".")
plt.plot(df[df["Ten-Year Age Groups"] == "55-64"]["Year"], df[df["Ten-Year Age Groups"] == "55-64"]["Crude Rate"], label = "55-64", marker = "1")
plt.plot(df[df["Ten-Year Age Groups"] == "65-74"]["Year"], df[df["Ten-Year Age Groups"] == "65-74"]["Crude Rate"], label = "65-74", marker = "8")
plt.plot(df[df["Ten-Year Age Groups"] == "75-84"]["Year"], df[df["Ten-Year Age Groups"] == "75-84"]["Crude Rate"], label = "75-84", marker = "x")
plt.plot(df[df["Ten-Year Age Groups"] == "85+"]["Year"], df[df["Ten-Year Age Groups"] == "85+"]["Crude Rate"], label = "85+", marker = "|")
plt.xlabel("Years")
plt.ylabel("Crude Death Rate")
plt.title("Years vs Crude Death Rate from 1999 - 2020")
plt.legend(title = "Age Group", loc = "upper left")
```

```python
# 1)
initial = 100
best = 0.01676
medium = 0.00549
worst = -0.045000
years = np.arange(1,11)


def f(n, r):
    return ((r + 1) ** n) * initial

plt.plot(years, f(years, best), label = "Best Growth Rate", marker = "o")
plt.plot(years, f(years, medium), label = "Medium Growth Rate", marker = "|")
plt.plot(years, f(years, worst), label = "Worst Growth Rate", marker = "x")

plt.xlabel("Years")
plt.ylabel("Population")
plt.title("Population Change Over Time (Years)")
plt.legend(title = "Growth Rates")
```
