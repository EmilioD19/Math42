### Reducing Population (Remove 1, Remove 5, Remove 1%, Remove 5%)

#### 10 Year Period

```python
# Set the natural growth rate
r_best = 0.01676

initial_population = 100

X = initial_population
x_k_one = initial_population
x_k_five = initial_population
x_k_5perc = initial_population
x_k_1perc = initial_population

h_one = -1
h_five = -5


N_year = 10

data = []
row = {"Time": 0,
       "X": X,
       "X_kill_one": x_k_one,
       "X_kill_5": x_k_five,
       "X_kill_one_perc": x_k_1perc,
       "X_kill_five_perc": x_k_5perc
       }

data.append(row)

for k in range(1,N_year):
    X = (1 + r_best) * X
    x_k_one = (1 + r_best) * x_k_one + h_one
    x_k_five = (1 + r_best) * x_k_five + h_five
    x_k_1perc = (1 + r_best) * x_k_1perc - (0.01 * x_k_1perc)
    x_k_5perc = (1 + r_best) * x_k_5perc - (0.05 * x_k_5perc)
    
    row = { "Time" : k,
           "X": X,
           "X_kill_one": x_k_one,
           "X_kill_five": x_k_five,
           "X_kill_one_perc": x_k_1perc,
           "X_kill_five_perc": x_k_5perc
    }
    data.append(row)
    
df = pd.DataFrame(data)
    
    
plt.plot(df["Time"], df["X"], "x", label = "natural")
plt.plot(df["Time"], df["X_kill_one"], "|", label = "kill 1 annually")
plt.plot(df["Time"], df["X_kill_five"], "-", label = "kill 5 annually")
plt.plot(df["Time"], df["X_kill_one_perc"], "--", label = "kill one percent annually")
plt.plot(df["Time"], df["X_kill_five_perc"], "o", label = "kill five percent annually")

plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Strategies for Population Trimming")
plt.legend(title = "Population Strategies")
    
```

### Stabalizing the population

```python
''' 
model for logistic growth
X(n) = X(n-1) + r(1 - (X(n-1) / K) * X(n - 1)

f(x) = x + rx(1 - x / K)

Assume K = 200, r = 0.01676, Let g be the fixed point of the system.
Then, g = g + rg(1 - g / K) or 0 = rg (K - g) so g = 0 and 200.

Testing the stability of the model we get: f' = 1 + r - 2rx / K.
Thus, f'(200) = 0.983
Since 0.983 < 1, then g = 200 is stable


'''

r_best = 0.01676

initial_population = 100

X = initial_population
carrying_cap = 200


N_year = 1000

data = []
row = {"Time": 0,
       "X": X
       }

data.append(row)

for k in range(1,N_year):
    X = X + r_best * (1 - (X / carrying_cap)) * X
    
    row = { "Time" : k,
           "X": X,
    }
    data.append(row)
    
df = pd.DataFrame(data)
    
    
plt.plot(df["Time"], df["X"], "x", label = "natural")

plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Logistic Growth Model")
plt.legend(title = "Population Strategies")
    
```

I created a logistic growth model. By setting the carrying capacity to 200, and changing the model so that the population growth decreases over time, we can see that the population will naturally decline over the course of several hundred years.

### Strategies that Stabilize at 50 and 200 


```python
# Suppose I want the population to stabalize at 50 bobcats
# 50 = -h / r, Since r = -0.045, then h = 2.25
# For a stable population of 200, h = 9

# Set the natural growth rate
r_worst = -0.045

initial_population = 100

X = initial_population

# 50 population
X_stable_50 = initial_population
h_stable_50 = 2.25

# 200 population
X_stable_200 = initial_population
h_stable_200 = 9


N_year = 100

data = []
row = {"Time": 0,
       "X": X,
       "X_stable_50": X_stable_50,
       "X_stable_200": X_stable_200
       }

data.append(row)

for k in range(1,N_year):
    X = (1 + r_best) * X
    X_stable_50 = (1 + r_worst) * X_stable_50 + h_stable_50
    X_stable_200 = (1 + r_worst) * X_stable_200 + h_stable_200
    
    row = { "Time" : k,
           "X": X,
           "X_stable_50": X_stable_50,
           "X_stable_200": X_stable_200
           }
    data.append(row)
    
df = pd.DataFrame(data)
    
    
plt.plot(df["Time"], df["X"], "x", label = "natural")
plt.plot(df["Time"], df["X_stable_50"], "-", label = "stable 50")
plt.plot(df["Time"], df["X_stable_200"], "-", label = "stable 200")
plt.axhline(y = 50, label = "50 population", linestyle = "--")
plt.axhline(y = 200, label = "200 population", linestyle = "--")

plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Strategies for Population Growth")
plt.legend(title = "Annual Population Strategies")
    
```

### Affine Functions

Based off the given functions from Parts 3 and 5 the affine functions are:


1. h = -1
2. h = -5
3. h = 3
4. h = 10

Thus, the fixed point for each strategy can be identified. 

Given the fixed point is represented by: x = -h / r, each strategy has the following fixed point.

1. h = -1, r = 0.01676, fixed point = 59.66
2. h = -5, r = 0.01676 fixed point = 298.32
3. h = 3, r = - 0.045 fixed point = 66.6
4. h = 10, r = - 0.045 fixed point = 222.22

Overall, we identify these strategies to be affine because of the constant h factor. For the other strategies, the h changed in response to the previous population. Because of this, we do not consider these strategies to be affine. As a result, the following strategies are considered affine and have the associated fixed point.

1. kill one bobcat annually, 59.66
2. kill five bobcats annually, 298.32
3. add 3 bobcats annually, 66.6
4. add 10 bobcats annually, 222.22
